```python
import pandas as pd
```

```python
df = pd.read_csv('imigrantes_canada.csv')
```

```python
df.head()
```

```python
df.info()
```

Setar o índice como país sem criar um novo DF

```python
df.set_index('País', inplace=True)
```

Criando os intervalo de anos

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
argentina = df.loc['Argentina', anos]
```

Anos X Quantidade de Imigrantes

```python
brasil
```

- O código cria um DataFrame brasil_dados que tem duas colunas: 'ano' e 'imigrantes'.
- A coluna 'ano' contém os valores do índice da série brasil.
- A coluna 'imigrantes' contém os valores da série brasil.

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
brasil_dados = pd.DataFrame(brasil_dict)

argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
argentina_dados = pd.DataFrame(argentina_dict)
```

```python
brasil_dados
```

```python
import matplotlib.pyplot as plt
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao = df_comparacao.T

df_comparacao
```

```python
plt.figure(figsize=(8,4))

plt.plot(brasil_dados['ano'], brasil_dados['imigrantes'])

plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')

# Definindo os intervalos
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show() # Finalizando as confs do gráfico
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.title('Imigração do Brasil X Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
